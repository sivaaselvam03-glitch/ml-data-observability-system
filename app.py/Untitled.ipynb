```
import streamlit as st
import pandas as pd


from ingestion.data_ingestion import load_data
from quality_checks import run_quality_checks
from metrics_generator import generate_metrics
from anomaly_detector import detect_anomalies


st.set_page_config(page_title="Data Observability Platform", layout="wide")


menu = st.sidebar.selectbox(
    "Navigation",
    ["Home", "Upload Data", "Run Pipeline", "Results", "Dashboard"]
)


if "data" not in st.session_state:
    st.session_state.data = None

if "results" not in st.session_state:
    st.session_state.results = None


if menu == "Home":
    st.title("📊 ML Data Observability Platform")
    st.write("""
    This system allows users to:
    - Upload datasets
    - Run automated data quality checks
    - Detect anomalies using ML
    - View insights in dashboard
    """)

elif menu == "Upload Data":
    st.title("📂 Upload Dataset")

    file = st.file_uploader("Upload CSV file", type=["csv"])

    if file:
        df = pd.read_csv("data_observability.csv")
        st.session_state.data = df

        st.success("Data uploaded successfully!")
        st.dataframe(df.head())


elif menu == "Run Pipeline":
    st.title("⚙️ Run Data Pipeline")

    if st.session_state.data is None:
        st.warning("Please upload data first.")
    else:
        if st.button("Run Full Pipeline"):
            df = st.session_state.data

            
            qc_results = run_quality_checks(df)

            
            metrics = generate_metrics(df)

            
            anomalies = detect_anomalies(df)

            
            st.session_state.results = {
                "qc": qc_results,
                "metrics": metrics,
                "anomalies": anomalies
            }

            st.success("Pipeline executed successfully!")


elif menu == "Results":
    st.title("📈 Results")

    if st.session_state.results is None:
        st.warning("Run the pipeline first.")
    else:
        results = st.session_state.results

        st.subheader("🔍 Quality Check Results")
        st.write(results["qc"])

        st.subheader("📊 Metrics")
        st.write(results["metrics"])

        st.subheader("🚨 Anomalies")
        st.write(results["anomalies"])


elif menu == "Dashboard":
    st.title("📊 Power BI Dashboard")

    power_bi_url = "PASTE_YOUR_POWER_BI_EMBED_LINK_HERE"

    st.components.v1.iframe(power_bi_url, height=600)
```
